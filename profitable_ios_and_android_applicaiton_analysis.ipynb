# Removing duplicates 

```python
instagram_duplicates_android = []

for app in android:
    name = app[0]
    if name == 'Instagram':
        print(app)
        instagram_duplicates_android.append(name)

print('\nNumber of Instagram duplicates: ', len(instagram_duplicates_android))
```

The above code shows an example of the data we are usinng for this analysis posessing duplicate entries that need to be cleaned. 

In the case above, Instagram is found four times in the dataset. Each Instagram entry is the same EXCEPT for the review column. 

For the purposes of this analysis we will use the review column containig the most amount of reviews, and remove all other duplicate entries.

```python
def find_unique_and_duplicate(dataset, app_column, review_column):
    unique = {}
    duplicate = {}
    for app in dataset:
        name = app[app_column]
        if name in unique:
            duplicate[name] = app[review_column]
        else:
            unique[name] = app[review_column]
    return unique, duplicate 
```

The above function can find unique and duplicate entries in csv data sets, and select one column of choice to add values into a dictionary. For purposes of this project we will choose the review column to make sure we have app with the most reviews to ensure we are selecting the duplicate entries with the most up to date information. 

--- still need a way to ensure app with most recent information is the one in the unique dictionary that will work universally for datasets 

```python
ios_unique, ios_duplicate = find_unique_and_duplicate(ios, 1, 6)
print('Number of duplicate ios apps: ', len(ios_duplicate))
print('Number of unique ios apps: ', len(ios_unique))

print('\n')

android_unique, android_duplicate = find_unique_and_duplicate(android, 0, 3)
print('Number of duplicate android apps: ', len(android_duplicate))
print('Number of unqiue android apps: ', len(android_unique))
```

The above code deletes a problematic row from the dataset which has unclean data for the application type (1.9) and the rating column with '3.0M' --> this needs to be a float. 


```python
def find_most_reviewed(dataset, review_column):
    reviews_max = {}
    for app in dataset[1:]:
        name = app[0]
        n_reviews = float(app[review_column])
        if name in reviews_max and reviews_max[name] < n_reviews:
            reviews_max[name] = n_reviews
        elif name not in reviews_max:
            reviews_max[name] = n_reviews
    return reviews_max
```

```python
most_reviewed_android = find_most_reviewed(android, 3)
most_reviewed_ios = find_most_reviewed(ios, 5)

print('Most reviews list of android apps: ', len(most_reviewed_android))
print('Total android apps: ', len(android))

print('Most reviews list of ios apps: ',len(most_reviewed_ios))
print('Total ios apps: ', len(ios))
```

```python
ios_clean = [] 
ios_already_added = [] 

android_clean = [] 
android_already_added = [] 

for app in android[1:]: 
    name = app[0]
    n_reviews = float(app[3])
    
    if (most_reviewed_android[name] == n_reviews) and (name not in android_already_added):
        android_clean.append(app)
        android_already_added.append(name)
print('clean android data\n')
explore_data(android_clean, 0, 3, True)

print('\n\n\n')

print('clean ios data\n')
for app in ios[1:]: 
    name = app[0]
    n_reviews = float(app[5])
    
    if (most_reviewed_ios[name] == n_reviews) and (name not in ios_already_added):
        ios_clean.append(app)
        ios_already_added.append(name)
        
explore_data(ios_clean, 0, 3, True)
```

Sometimes even english applications will use special characters and emoji's is their names. We want to adjust the function to make sure we are including these english applications by making the 'is_english' tolerate a maximum of 3 non-english characters per app name. 

```python
def is_english(string):
    
    non_ascii = 0
    
    for character in string:
        if ord(character) > 127:
            non_ascii += 1   
        
    if non_ascii > 3:
        return False
    else: 
        return True 
  
print(is_english('Instagram'))
print(is_english('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
print(is_english('Docs To Go‚Ñ¢ Free Office Suite'))
print(is_english('Instachat üòú'))
```

We can use 'prime_genre' (ios[11] == ios[-5]) for ios applications. 

We can use 'Genres' (android[9] == android[-4]) and 'Category' (android[1]) for android applications. 

```python
def freq_table(dataset, index):
    
    table = {}
    total = 0
    
    for row in dataset:
        total += 1 
        value = row[index]
        if value in table:
            table[value] += 1 
        else:
            table[value] = 1 
            
    table_percentages = {}
    
    for key in table: 
        percentage = (table[key] / total) * 100 
        table_percentages[key] = percentage 
    
    return  table_percentages 



def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
print("Frequency table of the prime gennre of free english ios apps from highest to lowest\n")

display_table(ios_final, 11)

print("\nFrequency table of the category and genre of free english android apps from highest to lowest")
print("\nCategory")

display_table(android_final, 1)

print("\nGenre")
display_table(android_final, 9)
```

To make the data easier to work with we will turn the data into a flaot and remove commas from the data. 

```python
categories_android = freq_table(android_final, 1)

for category in categories_android: 
    total = 0 
    len_category = 0 
    for app in android_final: 
        category_app = app[1]
        if category_app == category: 
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            total += float(n_installs)
            len_category += 1 
    avg_n_installs = total / len_category
    print(category, ':', avg_n_installs)
            
```
